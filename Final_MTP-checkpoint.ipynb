# Linear regression

```python
def forward_feature_selection1(X, y):
    selected_features = []
    best_r2_adj = float('-inf')

    while True:
        best_feature = None
        best_r2_adj_feature = float('-inf')

        for feature in X.columns:
            if feature not in selected_features:
                features_subset = selected_features + [feature]
                X_subset = X[features_subset]

                # Handle missing values and infinite values
                X_subset = X_subset.replace([np.inf, -np.inf], np.nan)
                X_subset = X_subset.dropna()

                model1 = LinearRegression()
                model1.fit(X_subset, y)
                y_pred = model1.predict(X_subset)
                r2 = r2_score(y, y_pred)
                n = len(y)
                p = len(features_subset)
                r2_adj = 1 - ((1 - r2) * (n - 1) / (n - p - 1))

                if r2_adj > best_r2_adj_feature:
                    best_r2_adj_feature = r2_adj
                    best_feature = feature

        if best_feature is None:
            break

        selected_features.append(best_feature)
        best_r2_adj = best_r2_adj_feature

    return selected_features
```

```python
from sklearn.model_selection import GridSearchCV
def hyper_param_opt_model(X, y,model,param_grid):
    """
    hyper-paramerer optimize by GridSearch
    """
    gsc = GridSearchCV(
        estimator=model,
        param_grid=param_grid,
        cv=5, 
        scoring= 'neg_mean_squared_error',
        verbose=0,
        n_jobs=-1)
    
    grid_result = gsc.fit(X, y)
    best_params = grid_result.best_params_
    return best_params
```
