```python
def type_create(string):
    string1= string.split('-')
    if string1[0][0:2]=='80':
        return 'None'
    elif 'Sintered' in string1:
        return 'None'
    elif 'Cast' in string1:
        return 'Cast'
    else:
        return 'Wrought'
```

```python
def temp_create(tem):
    tem1= tem.split('-')
    if 'Cast' in tem1:
        return tem1[-3]
    else:
        return tem1[1]
```

```python
data2['tempers']=data2['name'].apply(temp_create)
```

```python
def comp_low(var):
    var = str(var)
    var1 = var.split(' to ')
    if len(var1)==2:
        return var1[0]
    else:
        return 0
def comp_high(va):
    va = str(va)
    va1 = va.split(' to ')
    if len(va1) == 2:
        return va1[1]
    else:
        return 0
```

```python
data3['Ag_low'] = data3['Ag'].apply(comp_low)
data3['Ag_high'] = data3['Ag'].apply(comp_high)
```

```python
data3['Al_low'] = data3['Al'].apply(comp_low)
data3['Al_high'] = data3['Al'].apply(comp_high)
data3['B_low'] = data3['B'].apply(comp_low)
data3['B_high'] = data3['B'].apply(comp_high)
data3['Be_low'] = data3['Be'].apply(comp_low)
data3['Be_high'] = data3['Be'].apply(comp_high)
data3['Bi_low'] = data3['Bi'].apply(comp_low)
data3['Bi_high'] = data3['Bi'].apply(comp_high)
data3['Co_low'] = data3['Co'].apply(comp_low)
data3['Co_high'] = data3['Co'].apply(comp_high)
data3['Cr_low'] = data3['Cr'].apply(comp_low)
data3['Cr_high'] = data3['Cr'].apply(comp_high)
data3['Cu_low'] = data3['Cu'].apply(comp_low)
data3['Cu_high'] = data3['Cu'].apply(comp_high)
data3['Fe_low'] = data3['Fe'].apply(comp_low)
data3['Fe_high'] = data3['Fe'].apply(comp_high)
data3['Ga_low'] = data3['Ga'].apply(comp_low)
data3['Ga_high'] = data3['Ga'].apply(comp_high)
data3['Li_low'] = data3['Li'].apply(comp_low)
data3['Li_high'] = data3['Li'].apply(comp_high)
data3['Mg_low'] = data3['Mg'].apply(comp_low)
data3['Mg_high'] = data3['Mg'].apply(comp_high)
data3['Mn_low'] = data3['Mn'].apply(comp_low)
data3['Mn_high'] = data3['Mn'].apply(comp_high)
data3['Ni_low'] = data3['Ni'].apply(comp_low)
data3['Ni_high'] = data3['Ni'].apply(comp_high)
data3['Pb_low'] = data3['Pb'].apply(comp_low)
data3['Pb_high'] = data3['Pb'].apply(comp_high)
data3['Si_low'] = data3['Si'].apply(comp_low)
data3['Si_high'] = data3['Si'].apply(comp_high)
data3['Sn_low'] = data3['Sn'].apply(comp_low)
data3['Sn_high'] = data3['Sn'].apply(comp_high)
data3['Ti_low'] = data3['Ti'].apply(comp_low)
data3['Ti_high'] = data3['Ti'].apply(comp_high)
data3['V_low'] = data3['V'].apply(comp_low)
data3['V_high'] = data3['V'].apply(comp_high)
data3['Zn_low'] = data3['Zn'].apply(comp_low)
data3['Zn_high'] = data3['Zn'].apply(comp_high)
data3['Zr_low'] = data3['Zr'].apply(comp_low)
data3['Zr_high'] = data3['Zr'].apply(comp_high)
```
